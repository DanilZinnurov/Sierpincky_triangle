```python
from matplotlib import pyplot as plt
from IPython.display import clear_output
import numpy as np
import seaborn as sns
n = 20000
x = [-10, 0, 10, -10, 1]
y = [0, 10, 0, 0, 1]   
sns.set (rc={"figure.figsize":( 20 , 10 )})
for i in range(n):
    clear_output(wait=True)
    n = np.random.random()
    if n <= 1/3:
        ver = 1
    elif n <= 2/3:
        ver = 2
    else:
        ver = 3
    P = [(x[-1] + x[ver - 1])/2, (y[-1] + y[ver-1])/2]
    x.append(P[0])
    y.append(P[1])
    sns.scatterplot(x = x, y = y, size = 0.1)
    plt.show()
```
